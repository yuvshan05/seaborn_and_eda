```python
x  = [1, 2, 3, 4, 5]
y1 = [50, 40, 70, 80, 20]
y2 = [80, 20, 20, 50, 60]
y3 = [70, 20, 60, 40, 60]
y4 = [80, 20, 20, 50, 60]

pic=sns.lineplot(x=x,y=y1,label='enfield',color="gray")
sns.lineplot(x=x,y=y2,label='honda')
sns.lineplot(x=x,y=y3,label='maruti')
sns.lineplot(x=x,y=y4,label='yamaha')
pic.set_title('sales')
pic.set_ylabel('sales in millions')
pic.set_xlabel('moths')
```
